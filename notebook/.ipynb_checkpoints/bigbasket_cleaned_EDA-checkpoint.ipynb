```python
import pandas as pd

df = pd.read_csv(r"C:\Users\geetb\Desktop\bigbasket-products-analysis\data\BigBasket Products.csv")

df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.isna().sum()
```

```python
df.columns = df.columns.str.lower().str.strip().str.replace(" " , "_")
df.head()
```

```python
print("Duplicates row:", df.duplicated().sum())
```

```python
df = df[df["product"].notna()]
df = df[df["brand"].notna()]
```

```python
df["rating"].unique()
```

```python
df["rating"] = df["rating"].fillna("Not Rated")
```

```python
df = df[df["description"].notna()]
```

```python
df[["sale_price", "market_price"]].isna().sum()
```

```python
df["sale_price"] = pd.to_numeric(df["sale_price"], errors="coerce")
df["market_price"] = pd.to_numeric(df["market_price"], errors="coerce")
```

```python
df.to_csv("bigbasket_products_cleaned.csv", index=False)
```

```python
df["discount_percent"] = ((df["market_price"] - df["sale_price"]) / df["market_price"]) * 100
df["discount_percent"] = df["discount_percent"].round(2)  
```

```python
df[["product", "sale_price", "market_price", "discount_percent"]].head(10)
```

# Which products have the highest discount?

To understand where the biggest discounts are happening, let's visualize the top 10 products with the highest percentage off from their market price.


```python
import matplotlib.pyplot as plt
import seaborn as sns

top_discounted = df.sort_values(by="discount_percent", ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(
    y=top_discounted["product"],
    x=top_discounted["discount_percent"],
    palette="viridis"
)

plt.xlabel("Discount Percentage")
plt.ylabel("Product")
plt.title("Top 10 Most Discounted Products")
plt.tight_layout()
plt.show()
```

# Which categories offer the highest average discount?

Let’s analyze the average discount % given across different product categories to understand which sections are offering the best deals.


```python
category_discount = df.groupby("category")["discount_percent"].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(
    x=category_discount.values,
    y=category_discount.index,
    palette="coolwarm"
)

plt.xlabel("Average Discount (%)")
plt.ylabel("Category")
plt.title("Top 10 Categories by Average Discount")
plt.tight_layout()
plt.show()
```

# Which brands are most premium or most discounted?

In this step, we explore brand-level pricing patterns to identify which brands are positioned as premium and which ones are giving large discounts.



```python
brand_price = df.groupby("brand")["market_price"].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(
    x=brand_price.values,
    y=brand_price.index,
    palette="flare"
)

plt.xlabel("Average Market Price")
plt.ylabel("Brand")
plt.title("Top 10 Most Premium Brands")
plt.tight_layout()
plt.show()
```

# Which brands are most premium or offer the highest discounts?

Now that we’ve explored category-level pricing, it’s important to understand how different brands are positioned in the market. 
These insights can help companies decide which brands are high-value and which ones might be leveraging discounts to drive sales.


```python
brand_discount = df.groupby("brand")["discount_percent"].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(
    x=brand_discount.values,
    y=brand_discount.index,
    palette="crest"
)

plt.xlabel("Average Discount %")
plt.ylabel("Brand")
plt.title("Top 10 Brands Giving Highest Average Discount")
plt.tight_layout()
plt.show()
```

# What is the distribution of products across categories?

To understand BigBasket’s product spread, we’ll create a pie chart showing how products are distributed across the top 10 categories.


```python
category_counts = df["category"].value_counts().head(10)


plt.figure(figsize=(8,8))
plt.pie(
    category_counts,
    labels=category_counts.index,
    autopct="%1.1f%%",
    startangle=140,
    colors=sns.color_palette("pastel"),
    wedgeprops={"edgecolor":"black"}
)

plt.title("Top 10 Categories by Product Count")
plt.axis("equal")  
plt.tight_layout()
plt.show()
```

# Which categories offer the highest average discounts?

This chart helps us understand which product categories are being aggressively discounted. These categories may represent fast-moving items, promotional focus, or areas of price war among brands.


```python
category_discount = df.groupby("category")["discount_percent"].mean().sort_values(ascending=False).head(10)


plt.figure(figsize=(8,8))
plt.pie(
    category_discount,
    labels=category_discount.index,
    autopct="%1.1f%%",
    startangle=90,
    colors=sns.color_palette("cool"),
    wedgeprops={"edgecolor":"black"}
)

plt.title("Top 10 Categories by Average Discount %")
plt.axis("equal")
plt.tight_layout()
plt.show()
```

# What are the rating trends for products?

Understanding how customers rate products is essential to assess product quality and satisfaction. Let's explore the distribution of ratings and see which ranges are most common.


```python
df["rating"] = pd.to_numeric(df["rating"], errors="coerce")
```

```python
df_rating = df[df["rating"].notna()]
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df_rating["rating"], bins=20, kde=True, color="orange")

plt.xlabel("Product Rating")
plt.ylabel("Number of Products")
plt.title("Distribution of Product Ratings")
plt.tight_layout()
plt.show()
```

# Top 10 Highest-Rated Products

Let’s now discover the highest-rated products in the BigBasket dataset. These may indicate customer favorites and high-quality offerings.



```python
top_rated = df_rating.sort_values(by="rating", ascending=False)[["product", "brand", "category", "sale_price", "rating"]].head(10)

top_rated.reset_index(drop=True, inplace=True)
top_rated
```

# Which categories have the highest average product ratings?

This chart shows which product categories customers are most satisfied with, based on average ratings. It can help identify which segments are consistently delivering quality.



```python
category_rating = df_rating.groupby("category")["rating"].mean().sort_values(ascending=False).head(10)

# Plot
plt.figure(figsize=(10,6))
sns.barplot(
    x=category_rating.values,
    y=category_rating.index,
    palette="magma"
)

plt.xlabel("Average Rating")
plt.ylabel("Category")
plt.title("Top 10 Categories by Average Product Rating")
plt.tight_layout()
plt.show()
```

# Is there any relationship between discount and rating?

We’ll now visualize whether products that offer bigger discounts also tend to get higher ratings, or if there's no clear relationship between the two.



```python
plt.figure(figsize=(10,6))

sns.scatterplot(
    data=df_rating,
    x="discount_percent",
    y="rating",
    alpha=0.5,
    color="orange"
)

plt.xlabel("Discount Percentage")
plt.ylabel("Product Rating")
plt.title("Do Higher Discounts Lead to Better Ratings?")
plt.grid(True)
plt.tight_layout()
plt.show()
```

There’s no clear correlation between discount percentage and product ratings.
This suggests that customers likely rate products based on their quality and satisfaction, not just because they got it cheaper.
